```python
# Probability and Statistics Questions

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform

## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

"""
The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. For example, if we roll a fair six-sided die, the PMF of rolling a specific number (say 3) is 1/6.

The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable to take on a particular value. For example, the PDF of a standard normal distribution is given by the function f(x) = (1/sqrt(2*pi)) * exp(-0.5 * x^2).
"""

# Example of PDF for a normal distribution
x = np.linspace(-3, 3, 1000)
pdf = norm.pdf(x, 0, 1)

plt.plot(x, pdf, label='PDF')
plt.title('Probability Density Function of a Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

```

```python
## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

"""
The Cumulative Density Function (CDF) of a random variable X is the function F(x) = P(X  x). It gives the probability that the random variable X is less than or equal to x. The CDF is useful because it provides a way to describe the distribution of probabilities up to a certain point.
"""

# Example of CDF for a normal distribution
cdf = norm.cdf(x, 0, 1)

plt.plot(x, cdf, label='CDF')
plt.title('Cumulative Density Function of a Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()

```

```python
## Q7: Explain uniform Distribution with an example.

"""
A uniform distribution is a type of probability distribution in which all outcomes are equally likely. For example, the roll of a fair die has a uniform distribution because each face of the die is equally likely to come up.

Example:
The PDF of a uniform distribution between a and b is f(x) = 1 / (b - a) for a  x  b.
"""

# Example of Uniform distribution
a, b = 0, 10
uniform_dist = uniform(a, b-a)
x = np.linspace(a, b, 1000)
pdf = uniform_dist.pdf(x)

plt.plot(x, pdf, label='Uniform PDF')
plt.title('Probability Density Function of a Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

```
