```python
# Q1: Plots and Their Uses in Seaborn
# Seaborn is a powerful library for statistical data visualization in Python. Here are five common plots and their uses:

# Scatter Plot

# Uses: To visualize the relationship between two continuous variables. Useful for identifying patterns or correlations.
# Line Plot

# Uses: To display trends over time or across categories. It's effective for showing continuous data points connected by straight lines.
# Box Plot

# Uses: To summarize the distribution of a continuous variable and identify outliers. It shows the median, quartiles, and potential outliers.
# Bar Plot

# Uses: To compare categorical data. It shows the relationship between a categorical variable and a numerical variable.
# Heatmap

# Uses: To visualize matrix-like data where individual values are represented by color. It's useful for correlation matrices, confusion matrices, etc.

```

```python
# Q2: Line Plot with "fmri" Dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri = sns.load_dataset("fmri")

# Plotting line plot using seaborn
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Adding title and labels
plt.title('Signal vs Timepoint for different events and regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

```python
# Q3: Box Plots with "titanic" Dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")

# Plotting box plots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age distribution across Passenger Class')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare distribution across Passenger Class')

plt.tight_layout()
plt.show()
```

```python
# Q4: Histogram with "diamonds" Dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset("diamonds")

# Plotting histogram
plt.figure(figsize=(8, 5))
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)
plt.title('Histogram of Price with Cut Categories')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()




```

```python
# Q5: Pair Plot with "iris" Dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset("iris")

# Plotting pair plot
sns.pairplot(iris, hue="species")
plt.title('Pair Plot of Iris Dataset')
plt.show()

```

```python
# Q6: Heatmap with "flights" Dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset("flights")

# Reshape the data for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
