```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

# Class: A blueprint or template for creating objects. It defines the attributes (data) and methods (functions) that objects of that class will have.
# Object: An instance of a class. It's a concrete entity with its own state (values for its attributes) and behavior (defined by the class's methods).
# Example:

class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print("Woof!")

# Create two objects (instances) of the Dog class
my_dog = Dog("Fido", "Labrador")
your_dog = Dog("Buddy", "Golden Retriever")

```

```python

Here are the answers to your questions:

Q1. Classes and Objects in OOP:

Class: A blueprint or template for creating objects. It defines the attributes (data) and methods (functions) that objects of that class will have.
Object: An instance of a class. It's a concrete entity with its own state (values for its attributes) and behavior (defined by the class's methods).
Example:

Python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print("Woof!")

# Create two objects (instances) of the Dog class
my_dog = Dog("Fido", "Labrador")
your_dog = Dog("Buddy", "Golden Retriever")
Use code with caution. Learn more
Q2. Four Pillars of OOPs:

Encapsulation: Bundling data (attributes) and code (methods) together within a class, protecting data and controlling access to it.
Inheritance: Allowing new classes (subclasses) to inherit properties and behaviors from existing classes (parent classes), promoting code reuse and extensibility.
Polymorphism: Ability of objects of different classes to respond differently to the same method call, enabling flexible and adaptable code.
Abstraction: Focusing on essential features and hiding implementation details, simplifying code and making it easier to manage complexity.
Q3. init() Function:

The __init__() function is the constructor of a class.
It's automatically called when an object is created to initialize its attributes.
It's used to set the initial state of the object.
Example:

Python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
Use code with caution. Learn more
Q4. self in OOPs:

self refers to the current object instance within a class's methods.
It's used to access and modify the object's attributes and call its methods.
It's the first argument passed to every method in a class.
Q5. Inheritance:

It's the mechanism by which a new class (subclass) acquires the properties and behaviors of an existing class (parent class).
Types of inheritance:
Single Inheritance: A subclass inherits from one parent class.
Multiple Inheritance: A subclass inherits from multiple parent classes.
Multilevel Inheritance: A subclass inherits from a parent class, which itself inherits from another class.
Hierarchical Inheritance: Multiple subclasses inherit from a single parent class.
Example of Single Inheritance:

Python
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        print("Woof!")
```
