```python
# Q2: What is a scatter plot? Generate and plot a scatter plot using the provided data.
# Scatter Plot:
# A scatter plot is a type of plot that displays values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

# Code to generate and plot a scatter plot:

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Subplot Function:
# The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid layout within the same figure, which is useful for comparing multiple plots side by side.

# Code to draw four line plots using the subplot() function:


import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

```python
# Q4: What is a bar plot? Plot a bar plot and a horizontal bar plot using the provided data.
# Bar Plot:
# A bar plot is a chart that presents categorical data with rectangular bars. Each bar's length or height is proportional to the values it represents. It is used to compare different categories.

# Code to plot a bar plot and a horizontal bar plot:

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color=['blue', 'orange', 'green', 'red'])
plt.title('Company Profit Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

# Horizontal Bar Plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color=['blue', 'orange', 'green', 'red'])
plt.title('Company Profit Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

```python
# Q5: What is a box plot? Plot a box plot using the provided data.
# Box Plot:
# A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum"). It is used to identify the central tendency, dispersion, and skewness of the data.

# Code to plot a box plot:


import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=['Box1', 'Box2'])
plt.title('Box Plot of Generated Data')
plt.ylabel('Values')
plt.show()




```
